# Area under curve between two limits:

```python
import numpy as np
from scipy.interpolate import make_interp_spline
import matplotlib.pyplot as plt

x = np.array([-2, 0, 2, 4, 6, 8, 10])
y = np.array([25, 10, 5, 15, 18, 12, 17])
 
xy = make_interp_spline(x, y)

X = np.linspace(x.min(), x.max(), 500)
Y = xy(X)

plt.plot(X, Y)

plt.show()

def integrate(x, y):
    area = np.trapz(y=y, x=x)
    return area
print("Area under curve between two limits :",integrate(x, y))
```
